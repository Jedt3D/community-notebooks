```python
input_image_for_prediction = "./classes_dataset/original_images/001.png" # path to the image to predict
model = "user_id/segment_50ep" # model name on huggingface hub
```

Visualize prediction

```python
# load image_processor and model 
image = Image.open(input_image_for_prediction)
image_processor = AutoImageProcessor.from_pretrained("nvidia/mit-b0", reduce_labels=True)
model = AutoModelForSemanticSegmentation.from_pretrained(model)
inputs = image_processor(images=image, return_tensors="pt")
outputs = model(**inputs) # predict
logits = outputs.logits 
```

```python
# postprocess
upsampled_logits = nn.functional.interpolate(
    logits,
    size=image.size[::-1],
    mode="bilinear",
    align_corners=False,
)
```

```python
# plot the prediction as a segmentation map
pred_seg = upsampled_logits.argmax(dim=1)[0]

color_seg = np.zeros((pred_seg.shape[0], pred_seg.shape[1], 3), dtype=np.uint8)
palette = np.array(ade_palette())

for label, color in enumerate(palette):
    color_seg[pred_seg == label, :] = color
    color_seg = color_seg[..., ::-1]  # convert to BGR

img = np.array(image) * 0.5 + color_seg * 0.5  # plot the image with the segmentation map
img = img.astype(np.uint8)

plt.figure(figsize=(15, 10))
plt.imshow(img)
plt.show()
```

Using Pipeline for inference

```python
# use the pipeline to predict
image = Image.open(input_image_for_prediction)
segmenter = pipeline("image-segmentation", model=model)
segmenter(image , return_all_scores=True)
```
